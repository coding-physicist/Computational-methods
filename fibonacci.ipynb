```python
import random as rd
import math as mt
a=[[],[],[]]
for z in range(2,8):
 n=10**z
 count=0
 for i in range (n):
    x=rd.random()
    y=rd.random()
    if y<mt.sqrt(1-x**2):
        count+=1
 pi=count/n*4
 a[0].append(z)
 a[1].append(pi)
 a[2].append(abs(pi-mt.pi))
print(a[0])
print(a[1])
print(a[2])
```

```python
import numpy as np
means=[[],[]]
i=0
while i<8:
 def fnctn(x):
    y=-x**3+6*x**2-x+17
    return y
 def f(x):
    return -2+x*7
 x=list(map(f,np.random.rand(10**i)))
 result=list(map(fnctn,x))
 means[0].append(i)
 means[1].append(np.mean(result))
 i+=1
print(means)
```

```python
import numpy as np
import matplotlib.pyplot as plt
t=1
D=100
def fn(x):
    global t,D
    y=1/np.sqrt(4*D*t)*np.exp(-x**2/(2*D))
    return y
a=np.arange(-100,100,0.01)
for i in range (5):
 plt.plot(a,list(map(fn,a)),label=(r'$D={}$').format(D))
 D=D*2
plt.legend(fontsize=10)
plt.show()
data=[[a],[list(map(fn,a))]]
data.head()
```

```python
x,y=0,0
a=[1,-1,0]
table=[[],[]]
moves=[(0,1),(1,0),(-1,0),(0,-1)]
import matplotlib.pyplot as plt
import random as rd
import math as mt

posx=[0]
posy=[0]
i=0
while i<(200000):
    move=rd.choice(moves)
    x=posx[-1]+move[0]
    y=posy[-1]+move[1]

    posx.append(x)
    posy.append(y)
    i+=1
    rfinal=mt.sqrt(posx[-1]**2+posy[-1]**2)
    table[0].append(i)
    table[1].append(rfinal)

plt.hist(table[1],bins=40,edgecolor='#232323')
plt.show()
plt.plot(table[0],table[1])
plt.show()
plt.plot(posx,posy)
plt.show()
```
